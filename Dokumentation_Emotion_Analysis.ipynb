# Dokumentation zum Projekt Emotion Recognition

## 1. Übersicht und Setup


Das Ziel dieses Projekts war die Entwicklung eines "Add-Ons" für Videochat-Plattformen wie Zoom, das dem Benutzer sofortiges Feedback über die Emotionen der Meeting-Teilnehmer liefert.
***
Es wurden zwei Ansätze verfolgt. Einmal die Emotionserkennung "from scratch" durch Training eines CNNs von Grund auf sowie Emotionserkennung durch Nutzung der Toolbox PyFeat.

#### CNN-Ansatz

Code und Dokumentation: **Main Code _ CNN for Emotion Recognition.ipynb**

#### Toolbox-Ansatz

Es wurden zwei Code Varianten entwickelt *CodeBild* und *CodeArray*, jede mit Vor- und Nachteilen. Da diese sich sehr ähneln wird lediglich CodeArray ausführlich dokumentiert.

CodeBild: 
- Vorteile: Die Emotionserkennung wird kontinuierlich durchgeführt.
- Nachteile: Das Bild wird zunächst auf dem PC gespeichert.

CodeArray:
- Vorteile: Das Bild wird nicht auf dem PC gespeichert.
- Nachteile: Es gibt keinen automatischen Loop.

Code und Dokumentation: **CodeArray_Documentation.ipynb**

Code: **CodeArray.ipynb**

Code: **CodeBild.ipynb** 
  

## 1.1 Installation und Setup des Projekts

### a) Installiere Anaconda nach dieser Anleitung
https://docs.anaconda.com/free/anaconda/install/index.html

### b) Erstelle ein neues Environment in Anaconda Navigator

Im Reiter *Environment* über den Button *Create* ein neues Environment erstellen.



### c) Aktiviere das Environment in Anaconda Navigator






### d) Installation der benötigten Packages

Im Reiter *Environments* bei dem aktiven Environment auf das grüne Dreieck klicken.
Terminal öffnen mit *Open Terminal*:




Folgende Packages über das Terminal mit *pip install* installieren:

- matplotlib
- seaborn 
- scikit-learn
- keras
- tensorflow
- pandas
- numpy
- py-feat




### e) Jupiter Notebook ausführen


Im Reiter *Home* Jupiter Notebook öffnen über *Launch*.






In Jupiter Notebook zum Speicherort des Notebooks wechseln und öffnen.

Den Code ausführen durch Klicken auf *Run* Button.

## 1.3 Nützliche Befehle für das Anaconda-Terminal

### Auflistung aller mit *pip* oder *conda* installierten Packages:

	conda env export

	conda env export > my_conda_env.yml

	conda env create --file my_conda_env.yml
